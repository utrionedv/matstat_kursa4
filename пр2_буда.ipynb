```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame()
df2 = pd.DataFrame()
df1['1'] = pd.read_csv("/content/1.txt", header=None)
df2['2'] = pd.read_csv("/content/2.txt", header=None)
df1['3'] = pd.read_csv("/content/3.txt", header=None)
df2['4'] = pd.read_csv("/content/4.txt", header=None)
```

## –†—è–¥ 1

```python

print('1. –†–∞—Å—Å—á–∏—Ç–∞—Ç—å —Ç–æ—á–µ—á–Ω—ã–µ –æ—Ü–µ–Ω–∫–∏ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è:')
print()

mean1 = df1['1'].mean()
print('–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ 1 —Ä—è–¥–∞: ', mean1)
std1 = df1['1'].std(ddof = 1)
print('C—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: 1 —Ä—è–¥–∞: ', std1)

print()
print()
print('2. –ù–∞–π—Ç–∏ –≥—Ä–∞–Ω–∏—Ü—ã –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ –º–∞—Ç. –æ–∂–∏–¥–∞–Ω–∏—è –ø–æ –ø—Ä–∞–≤–∏–ª—É –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è, –∏—Å–ø–æ–ª—å–∑—É—è —Ç–∞–±–ª–∏—Ü—É –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Ñ—É–Ω–∫—Ü–∏–∏ –õ–∞–ø–ª–∞—Å–∞:')
print()

xy = 1.96
delta = xy * (std1 / (len(df1['1']) ** 0.5))
print('–ò–Ω—Ç–µ—Ä–≤–∞–ª –º–∞—Ç. –æ–∂–∏–¥–∞–Ω–∏—è 1 —Ä—è–¥–∞ (–ø–æ –ø—Ä–∞–≤–∏–ª—É –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è): [', mean1 - delta, '; ', mean1 + delta, ']')
# print(delta)

print()
print()
print('3. –ü–æ –ø—Ä–∞–≤–∏–ª—É t-—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –°—Ç—å—é–¥–µ–Ω—Ç–∞, –∏—Å–ø–æ–ª—å–∑—É—è —Ç–∞–±–ª–∏—Ü—É –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π ùë°ùõæ,ùëõ t-—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø—Ä–∏ –∑–Ω–∞—á–µ–Ω–∏–∏ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç–∏ ùõæ = 0.95:')
print()

t99 = 1.98
delta = t99 * (std1 / (len(df1['1']) ** 0.5))
print('–ò–Ω—Ç–µ—Ä–≤–∞–ª –º–∞—Ç. –æ–∂–∏–¥–∞–Ω–∏—è 1 —Ä—è–¥–∞ (–ø–æ –ø—Ä–∞–≤–∏–ª—É t-—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –°—Ç—å—é–¥–µ–Ω—Ç–∞): [', mean1 - delta, '; ', mean1 + delta, ']')
# print(delta)

print()
print()
print('–ù–∞–π—Ç–∏ –≥—Ä–∞–Ω–∏—Ü—ã –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ –¥–ª—è —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –ø–æ –æ—Ü–µ–Ω–∫–µ ùúí2-—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø—Ä–∏ –∑–Ω–∞—á–µ–Ω–∏–∏ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç–∏ ùõæ = 0.95:')
print()
X99_1 = 128.42
X99_2 = 73.36

print('–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è 1 —Ä—è–¥–∞: [',
 (std1 * ((len(df1['1']) - 1) ** 0.5)) / (X99_1 ** 0.5), '; ', (std1 * ((len(df1['1']) - 1) ** 0.5)) / (X99_2 ** 0.5), ']')
```
